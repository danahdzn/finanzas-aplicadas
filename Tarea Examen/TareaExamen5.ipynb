```python
# -*- coding: utf-8 -*-
"""
Created on Wed Nov 22 11:34:35 2023

@author: Meva
"""

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import importlib
import random
import pickle

# importar el archivo de soporte
from google.colab import drive
drive.mount('/content/drive')
import sys
sys.path.append('/content/drive/MyDrive/ArchivosCSV/ArchivosAuxiliares/')
import market_data
import portfolio
import assert_frame

"""
Explicación de la Tarea-Examen 05.

Tu gerente te pide que le mejores el portafolios que ella ha elegigo.
El portafolios elegido por tu gerente es equi-weight o equi-ponderado.
Tú tienes que escoger un portafolios óptimo, es decir de Markowitz.

Inputs:
El único parámetro que puedes modificar es el target_return de Markowitz.

Output:
Tu gerente quiere que tu portafolios mejore el suyo en 3 criterios:
1. Mayor rendimiento, al menos 0.1%.
2. Menor volatilidad, al menos 0.1%.
3. Mejor cociente de Sharpe.

Sólo si logras mejorar los 3 criterios vas a tener éxito y bono asegurado:
nota = 10.0.

Si no logras mejorar los 3 criterios tu gerente te corre del fondo:
nota = 0.0.

"""


##############################
# datos que debes proporcionar
##############################
cuenta =317163027  # tu número de cuenta, como int
target_return = 0.111564 # tu rendimiento "mágico" para el portafolios de Markowitz


############
# evaluación
############
with open('/content/drive/MyDrive/ArchivosCSV/ArchivosAuxiliares/examen_05_instances.pickle','rb') as f:
    cuenta_rics = pickle.load(f)
rics = cuenta_rics[cuenta]
print('-----')
print('Para el número de cuenta ' + str(cuenta) + ' ésta es su lista de activos personalizada: ')
print('rics = ' + str(rics))
port_mgr = portfolio.manager(rics, 1)
port_mgr.compute_covariance()
port_equi_weight = port_mgr.compute_portfolio('equi-weight')
port_markowitz = port_mgr.compute_portfolio('markowitz', target_return)
count_success = 0
count_total = 3
improved_return = port_markowitz.return_annual - port_equi_weight.return_annual > 0.001
improved_volatility = port_markowitz.volatility_annual - port_equi_weight.volatility_annual < 0.001
improved_sharpe = port_markowitz.sharpe_ratio > port_equi_weight.sharpe_ratio
print('-----')
print('Checking if you improved the return: difference > 0.001')
print('port_equi_weight.return_annual = ' + str(port_equi_weight.return_annual))
print('port_markowitz.return_annual = ' + str(port_markowitz.return_annual))
if improved_return:
    count_success += 1
print('Improved return = ' + str(improved_return))
print('-----')
print('Checking if you improved the volatility: difference > 0.001')
print('port_equi_weight.volatility_annual = ' + str(port_equi_weight.volatility_annual))
print('port_markowitz.volatility_annual = ' + str(port_markowitz.volatility_annual))
if improved_volatility:
    count_success += 1
print('Improved volatility = ' + str(improved_return))
print('-----')
print('Checking if you improved the Sharpe ratio: difference > 0')
print('port_equi_weight.sharpe_ratio = ' + str(port_equi_weight.sharpe_ratio))
print('port_markowitz.sharpe_ratio = ' + str(port_markowitz.sharpe_ratio))
if improved_sharpe:
    count_success += 1
print('Improved Sharpe Ratio = ' + str(improved_return))
note = 10.0 if count_success == count_total else 0.0
print('-----')
print('Total improvements: ' +str(count_success) + ' of ' + str(count_total))
print('-----')
print('Your final note is ' + str(note))


```
